```python
import pandas as pd
import sklearn.datasets as datasets
import sklearn.svm as svm
import pylab as pl


iris = datasets.load_iris()
digits = datasets.load_digits()
diabetes = datasets.load_diabetes()
```

## Pandas representation of an sklearn dataset

```python
idf = pd.DataFrame(data=iris.data, index=iris.target, columns=iris.feature_names)

idf[::10]
```

```python
# Now let's improve it by adding the target_name and getting a unique index
idf = idf.reset_index()
idf["index"] = idf["index"].apply(lambda x: iris.target_names[x])
idf = idf.rename(columns={"index": "target_name"})

idf[::10]
```

## Plotting an sklearn dataset (scatter plot)

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Note that the 'sepal width (cm)' dimension is missing.
x_label = "petal width (cm)"
y_label = "sepal length (cm)"
z_label = "petal length (cm)"
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)
ax.set_zlabel(z_label)

for name, color in zip(idf.target_name.unique(), ("b", "r", "g")):
    flower = idf[idf["target_name"]==name]
    ax.scatter(flower[x_label],
               flower[y_label],
               flower[z_label],
               c=color, marker="o")

plt.show()
```

## Supervised learning example: Support Vector Classification

```python
training_data = digits.data[1:]
training_target = digits.target[1:]

test_datum = digits.data[0]

# Choose the estimator.
classifier = svm.SVC(gamma=0.001, C=100.)
# Learn from the training set.
classifier.fit(training_data, training_target)

classifier.predict(test_datum)
```

```python
# Now let's check if this is actually the number we should have predicted.
import pylab as pl 
pl.gray() 
pl.matshow(test_datum.reshape(8, 8)) 
pl.show() 
```

```python
# Yes it is!
```

## Unsupervised learning example: K-means clustering

```python
### K-means clustering ####
from sklearn import cluster

k_means = cluster.KMeans(n_clusters=3)
k_means.fit(iris.data)

print k_means.labels_
```

```python
# Compare with:
print iris.target
```

```python
# Hmm, this is not super clear, let's see if we can make a side by side comparison of the result:
comp = pd.DataFrame({"k_means_guess": k_means.labels_ , "real_group": iris.target})
target_to_target_names = lambda x: iris.target_names[x]
comp["k_means_guess"] = comp["k_means_guess"].apply(target_to_target_names)
comp["real_group"] = comp["real_group"].apply(target_to_target_names)

comp
```

```python
# We can still get a better view of our result by plotting it:
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_label = "petal width (cm)"
y_label = "sepal length (cm)"
z_label = "petal length (cm)"
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)
ax.set_zlabel(z_label)

icdf = pd.DataFrame(data=iris.data, index=k_means.labels_, columns=iris.feature_names)
icdf = icdf.reset_index()
icdf["index"] = icdf["index"].apply(lambda x: iris.target_names[x])
icdf = icdf.rename(columns={"index": "target_name"})


for name, color in zip(idf.target_name.unique(), ("b", "r", "g")):
    flower = icdf[icdf["target_name"]==name]
    ax.scatter(flower[x_label],
               flower[y_label],
               flower[z_label],
               c=color, marker="o")
    
plt.show()
```
